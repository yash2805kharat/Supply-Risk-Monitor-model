```
#Removing URLs
def remove_urls(text):
	url_pattern = r'https?://\S+|www\.\S+'
	without_urls = re.sub(pattern=url_pattern, repl='', string=text)
	return without_urls
```

```
#Removing emojis
def remove_emojis(text):
	emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)
	without_emoji = emoji_pattern.sub(r'',text)
	return without_emoji
```

```
#Cleaning unwanted texts
def clean_data(text):
    text=text.replace('âš ï','')
    text=text.replace('âœ','')
    text=text.replace('&amp;','')
    text=text.replace('â€™','')
    text=text.replace('â€œ','')
    text=text.replace('â€','')
    text=text.replace('ðŸ”¥ðŸš’','')
    text=text.replace('â„ï¸','')
    text=text.replace('â€¼ï¸','')
    text=text.replace('ðŸŽµðŸŽ¶','')
    text=text.replace('ðÿž¶ðÿžµ','')
    text=text.replace('ðÿžµðÿž¶','')
    text=text.replace('ðŸŽ¶ðŸŽµ','')
    text=text.replace('ðŸŽ¶ðŸª•','')
    text=text.replace('\n','')
    text=text.replace('ðŸ”‰','')
    text=text.replace('Â','')
    text=text.replace("''",'')
    return text
```

```
#Removing Extra Whitespaces
def remove_extra_spaces(text):
	space_pattern = r'\s+'
	without_space = re.sub(pattern=space_pattern, repl=" ", string=text)
	return without_space
```

```
#Removing punctuations using string library
def remove_punctuation(text):
	return text.translate(str.maketrans('', '', punctuation))
```

```
twt_weather_data['Cleaned_Text'] = twt_weather_data['Cleaned_Text'].apply(remove_urls)
twt_weather_data['Cleaned_Text'] = twt_weather_data['Cleaned_Text'].apply(remove_emojis)
twt_weather_data['Cleaned_Text'] = twt_weather_data['Cleaned_Text'].apply(clean_data)
twt_weather_data['Cleaned_Text'] = twt_weather_data['Cleaned_Text'].apply(remove_extra_spaces)
twt_weather_data['Cleaned_Text'] = twt_weather_data['Cleaned_Text'].apply(remove_punctuation)
```

```
#Removing Stopword using Spacy
en = spacy.load('en_core_web_sm')
stopwords = en.Defaults.stop_words
```

```
#Creating the list of stopwords
stop_word_list=list(stopwords)
print(stop_word_list)
```

```
#Removing few words that we would require for tweet classification
remove_word={'throughout','becomes','afterwards'}
#Creating a list after removing those few words
stop_word_list_new=[ele for ele in stop_word_list if ele not in remove_word]
```

```
#Removing Stop words from the Cleaned Text column
for a in range(twt_weather_data.shape[0]):
  text = twt_weather_data['Cleaned_Text'].iloc[a]
  words = [word for word in text.split() if word.lower() not in stop_word_list_new]
  twt_weather_data['Cleaned_Text'].iloc[a] = " ".join(words)
```

```
print(twt_weather_data['Text'].iloc[0],'\n')
print(twt_weather_data['Cleaned_Text'].iloc[0])
```

```
# Split into training and testing data
x = twt_weather_data['Cleaned_Text']
y = twt_weather_data['Check']
x, x_test, y, y_test = train_test_split(x,y, stratify=y, test_size=0.25, random_state=42)
```

```
#Converting the collection of text in cleaned text column into a matrix of token counts.
vec = CountVectorizer()
x = vec.fit_transform(x).toarray()
x_test = vec.transform(x_test).toarray()
```

```
#Applying Naives Bayes model for text classification
model = MultinomialNB()
model.fit(x, y)
```

```
#Predicting the target variable through model
pred=model.predict(x_test)
```

```
#Model evaluation score on test data
acc_score = metrics.accuracy_score(y_test,pred)
f1_score = metrics.f1_score(y_test,pred)
recall_score=metrics.recall_score(y_test,pred)
precision_score=metrics.precision_score(y_test,pred)
print('Total accuracy classification score: {}'.format(acc_score))
print('Total F1 classification score: {}'.format(f1_score))
print('Total recall classification score: {}'.format(recall_score))
print('Total precision classification score: {}'.format(precision_score))
```

```
#Confusion matrix
confusion_matrix(y_test,pred)
```

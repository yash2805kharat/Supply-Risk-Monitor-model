```
#Lower case conversion of text column
supplychain_twt_data['Cleaned_Text']=supplychain_twt_data['Text'].str.lower()

#Removing URLs
import re
def remove_urls(text):
	url_pattern = r'https?://\S+|www\.\S+'
	without_urls = re.sub(pattern=url_pattern, repl='', string=text)
	return without_urls

for i in range(supplychain_twt_data.shape[0]):
  urls=supplychain_twt_data['Cleaned_Text'].iloc[i]
  urls_result = remove_urls(urls)
  supplychain_twt_data['Cleaned_Text'].iloc[i]=urls_result
```

```
# Implementation of emoji removing

def remove_emojis(text):
	"""
	Result :- string without any emojis in it
	Input :- String
	Output :- String
	"""
	emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)

	without_emoji = emoji_pattern.sub(r'',text)
	return without_emoji


# example text for emoji removing technique
for j in range(supplychain_twt_data.shape[0]):
  ex_emoji = supplychain_twt_data['Cleaned_Text'].iloc[j]
  emoji_result = remove_emojis(ex_emoji)
  supplychain_twt_data['Cleaned_Text'].iloc[j]=emoji_result
```

```
#cleaning unwanted texts

supplychain_twt_data['Cleaned_Text'].replace(regex=True, to_replace =['¢','€', 'Ã', '¬', 'Ð', '±','½','©','•','¾', '§', '¥', '«', '¤', '–', 'œ', '¡', '”', '|', 'â', '™', 'Â', 'Î', '¿', 'µ', '´', '‡','»', 'Ž', '®', 'º', 'Ï', 'ƒ', '¶', '¹', '┬', 'á', 'Γ', 'Ç', 'Ö','Ë','œ','Ã','£','ð','Ÿ','š','ª','²','Ã','£','°','¿','Â','ž','&amp;','&amp','â','ã','~','J&amp;J','','˜','ÿ’','›','†','ðŸ’™ðŸ†ðŸ’™ðŸ†ðŸ’™','ÿ','“','Ÿ‘','Ÿ’','Ÿ','ï','—','ï','ï','ÿ‘','ÿ—ï','ÿ—“ï','³','ÿ“','·:','‹','ÿ‘','ÿ˜','¨','’','“','¸','‘','‰'], value='', inplace=True)

def clean_data(text):
    text=text.replace('âš ï','')
    text=text.replace('âœ','')
    text=text.replace('&amp;','')
    text=text.replace('â€™','')
    text=text.replace('â€œ','')
    text=text.replace('â€','')
    text=text.replace('ðŸ”¥ðŸš’','')
    text=text.replace('â„ï¸','')
    text=text.replace('â€¼ï¸','')
    text=text.replace('ðŸŽµðŸŽ¶','')
    text=text.replace('ðÿž¶ðÿžµ','')
    text=text.replace('ðÿžµðÿž¶','')
    text=text.replace('ðŸŽ¶ðŸŽµ','')
    text=text.replace('ðŸŽ¶ðŸª•','')
    text=text.replace('ªð','')
    text=text.replace('²ð','')
    text=text.replace('µð','')
    text=text.replace('¹ð','')
    text=text.replace('ºð','')
    text=text.replace('ºðÿ','')
    text=text.replace('¼ð','')
    text=text.replace('½ðÿš','')
    text=text.replace('ðÿ','')
    text=text.replace('ðÿœš','')
    text=text.replace('ðÿš','')
    text=text.replace('ðÿž','')
    text=text.replace('\n','')
    for char in ['²the', '³ð', 'áÿ', 'áÿƒáž', 'áž', 'áž¼ážšážˆáÿ', 'áž½áž', 'áž¾ážˆáÿ', 'ážÿáÿ', 'ážÿáž', 'ážšáž', 'ážšáž¹áž', 'ážšáž¼ážœáž', 'ážšáž¾áž', 'âš', 'âž', 'žƒ', 'ƒï']:
      text=text.replace('char','')
    return text

supplychain_twt_data['Cleaned_Text'] = supplychain_twt_data['Cleaned_Text'].apply(clean_data)
```

```
# Removing Extra Whitespaces

import re

def remove_extra_spaces(text):
	space_pattern = r'\s+'
	without_space = re.sub(pattern=space_pattern, repl=" ", string=text)
	return without_space
  
supplychain_twt_data['Cleaned_Text'] = supplychain_twt_data['Cleaned_Text'].apply(remove_extra_spaces)
```

```
# Implementation of removing punctuations using string library

from string import punctuation

def remove_punctuation(text):
	return text.translate(str.maketrans('', '', punctuation))


# example text for removing punctuations
for i in range(supplychain_twt_data.shape[0]):
  ex_punct = supplychain_twt_data['Cleaned_Text'].iloc[i]
  supplychain_twt_data['Cleaned_Text'].iloc[i] = remove_punctuation(ex_punct)
```

```
#Removing Stopword
import spacy
en = spacy.load('en_core_web_sm')
stopwords = en.Defaults.stop_words

print(len(stopwords))

stop_word_list=list(stopwords)

stop_word_list_new=stop_word_list

for a in range(supplychain_twt_data.shape[0]):
  text = supplychain_twt_data['Cleaned_Text'].iloc[a]
  words = [word for word in text.split() if word.lower() not in stop_word_list_new]
  supplychain_twt_data['Cleaned_Text'].iloc[a] = " ".join(words)
```

```
#def function checkaccuracy is to find the best model suitable for tweet classification
def checkaccuracy(dtrees):
  dtrees=dtrees.fit(X_train, Y_train)
  predictions=dtrees.predict(X_test)
  train=dtrees.score(X_train, Y_train)
  test=dtrees.score(X_test, Y_test)
  acc_score = metrics.accuracy_score(Y_test,predictions)
  f1_score = metrics.f1_score(Y_test,predictions)
  recall_score=metrics.recall_score(Y_test,predictions)
  precision_score=metrics.precision_score(Y_test,predictions)
  cm1 = confusion_matrix(Y_test,predictions)
  scores=abs(cm1[0,1]-cm1[1,0])
  return train,test,cm1,scores,acc_score,f1_score,recall_score,precision_score
```

```
#
x = supplychain_twt_data['Cleaned_Text']
y = supplychain_twt_data['Checks']
X_train, X_test, Y_train, Y_test = train_test_split(x,y, stratify=y, test_size=0.25, random_state=42)

# Vectorize text reviews to numbers
vec = CountVectorizer()
X_train = vec.fit_transform(X_train).toarray()
X_test = vec.transform(X_test).toarray()

#1. Decision tree
dtrees=DecisionTreeClassifier(criterion='entropy')
train1,test1,cm1,scores1,acc_score1,f1_score1,recall_score1,precision_score1=checkaccuracy(dtrees)

#2. Bagging classifier with base estimator as Decision tree
dtrees=BaggingClassifier(base_estimator=DecisionTreeClassifier(),n_estimators=18)
train2,test2,cm2,scores2,acc_score2,f1_score2,recall_score2,precision_score2=checkaccuracy(dtrees)

#3. BaggingClassifier
dtrees=BaggingClassifier(n_estimators=2)
train3,test3,cm3,scores3,acc_score3,f1_score3,recall_score3,precision_score3=checkaccuracy(dtrees)

#4. Random forest classifier
dtrees=RandomForestClassifier(n_estimators=30)
train4,test4,cm4,scores4,acc_score4,f1_score4,recall_score4,precision_score4=checkaccuracy(dtrees)

#5. Ada boost classifier with base estimator as Decision tree
dtrees=AdaBoostClassifier(base_estimator=DecisionTreeClassifier(),n_estimators=2)
train5,test5,cm5,scores5,acc_score5,f1_score5,recall_score5,precision_score5=checkaccuracy(dtrees)

#6. Ada boost classifier
dtrees=AdaBoostClassifier(n_estimators=2)
train6,test6,cm6,scores6,acc_score6,f1_score6,recall_score6,precision_score6=checkaccuracy(dtrees)

#7. Gradient boosting classifier
dtrees=GradientBoostingClassifier(n_estimators=2)
train7,test7,cm7,scores7,acc_score7,f1_score7,recall_score7,precision_score7=checkaccuracy(dtrees)

#8. KNeighbor classifier
dtrees=KNeighborsClassifier(n_neighbors=20)
train8,test8,cm8,scores8,acc_score8,f1_score8,recall_score8,precision_score8=checkaccuracy(dtrees)

#9. Naive bayes
dtrees=MultinomialNB()
train9,test9,cm9,scores9,acc_score9,f1_score9,recall_score9,precision_score9=checkaccuracy(dtrees)

#10. Gaussian Naive bayes
dtrees=GaussianNB()
train10,test10,cm10,scores10,acc_score10,f1_score10,recall_score10,precision_score10=checkaccuracy(dtrees)

#creating tempdf to store all model values
tempdf=pd.DataFrame(data=[[train1,test1,cm1[0,0],cm1[0,1],cm1[1,0],cm1[1,1],scores1,acc_score1,f1_score1,recall_score1,precision_score1],[train2,test2,cm2[0,0],cm2[0,1],cm2[1,0],cm2[1,1],scores2,acc_score2,f1_score2,recall_score2,precision_score2],[train3,test3,cm3[0,0],cm3[0,1],cm3[1,0],cm3[1,1],scores3,acc_score3,f1_score3,recall_score3,precision_score3],[train4,test4,cm4[0,0],cm4[0,1],cm4[1,0],cm4[1,1],scores4,acc_score4,f1_score4,recall_score4,precision_score4],[train5,test5,cm5[0,0],cm5[0,1],cm5[1,0],cm5[1,1],scores5,acc_score5,f1_score5,recall_score5,precision_score5],[train6,test6,cm6[0,0],cm6[0,1],cm6[1,0],cm6[1,1],scores6,acc_score6,f1_score6,recall_score6,precision_score6],[train7,test7,cm7[0,0],cm7[0,1],cm7[1,0],cm7[1,1],scores7,acc_score7,f1_score7,recall_score7,precision_score7],[train8,test8,cm8[0,0],cm8[0,1],cm8[1,0],cm8[1,1],scores8,acc_score8,f1_score8,recall_score8,precision_score8],[train9,test9,cm9[0,0],cm9[0,1],cm9[1,0],cm9[1,1],scores9,acc_score9,f1_score9,recall_score9,precision_score9],[train10,test10,cm10[0,0],cm10[0,1],cm10[1,0],cm10[1,1],scores10,acc_score10,f1_score10,recall_score10,precision_score10]],columns=['train','test','cm[0,0]','cm[0,1]','cm[1,0]','cm[1,1]','score','acc_score','f1_score','recall_score','precision_score'],index=['DecisionTreeClassifier(criterion=entropy)','BaggingClassifier(base_estimator=DecisionTreeClassifier(),n_estimators=18)','BaggingClassifier(n_estimators=2)','RandomForestClassifier(n_estimators=30)','AdaBoostClassifier(base_estimator=DecisionTreeClassifier(),n_estimators=2)','AdaBoostClassifier(n_estimators=2)','GradientBoostingClassifier(n_estimators=2)','KNeighborsClassifier(n_neighbors=20)','MultinomialNB()','GaussianNB()'])
```

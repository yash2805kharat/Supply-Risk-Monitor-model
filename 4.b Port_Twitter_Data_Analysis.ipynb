```
#Lower case conversion of text column
port_extreme_weather_tweets['Cleaned_Text']=port_extreme_weather_tweets['Text'].str.lower()

#Removing URLs
def remove_urls(text):
	url_pattern = r'https?://\S+|www\.\S+'
	without_urls = re.sub(pattern=url_pattern, repl='', string=text)
	return without_urls
  
#Removing Emojis
def remove_emojis(text):
	emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)

	without_emoji = emoji_pattern.sub(r'',text)
	return without_emoji

#Cleaning unwanted texts
def clean_data(text):
    text=text.replace('âš ï','')
    text=text.replace('âœ','')
    text=text.replace('&amp;','')
    text=text.replace('â€™','')
    text=text.replace('â€œ','')
    text=text.replace('â€','')
    text=text.replace('ðŸ”¥ðŸš’','')
    text=text.replace('â„ï¸','')
    text=text.replace('â€¼ï¸','')
    text=text.replace('ðŸŽµðŸŽ¶','')
    text=text.replace('ðÿž¶ðÿžµ','')
    text=text.replace('ðÿžµðÿž¶','')
    text=text.replace('ðŸŽ¶ðŸŽµ','')
    text=text.replace('ðŸŽ¶ðŸª•','')
    text=text.replace('\n','')
    text=text.replace('ðŸ”‰','')
    text=text.replace('Â','')
    return text

#Removing Extra Whitespaces
def remove_extra_spaces(text):
	space_pattern = r'\s+'
	without_space = re.sub(pattern=space_pattern, repl=" ", string=text)
	return without_space
  
#Removing punctuations using string library
def remove_punctuation(text):
	return text.translate(str.maketrans('', '', punctuation))
  

#Applying function
port_extreme_weather_tweets['Cleaned_Text'] = port_extreme_weather_tweets['Cleaned_Text'].apply(remove_urls)
port_extreme_weather_tweets['Cleaned_Text'] = port_extreme_weather_tweets['Cleaned_Text'].apply(remove_emojis)
port_extreme_weather_tweets['Cleaned_Text'] = port_extreme_weather_tweets['Cleaned_Text'].apply(clean_data)
port_extreme_weather_tweets['Cleaned_Text'] = port_extreme_weather_tweets['Cleaned_Text'].apply(remove_extra_spaces)
port_extreme_weather_tweets['Cleaned_Text'] = port_extreme_weather_tweets['Cleaned_Text'].apply(remove_punctuation)
```

```
#Converting the collection of text in cleaned text column into a matrix of token counts.
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
x = vec.fit_transform(x).toarray()
x_test = vec.transform(x_test).toarray()
```

```
#def function checkaccuracy is to find the best model suitable for tweet text classification.
def checkaccuracy(dtrees):
  dtrees=dtrees.fit(x, y)
  predictions=dtrees.predict(x_test)
  train=dtrees.score(x, y)
  test=dtrees.score(x_test, y_test)
  acc_score = metrics.accuracy_score(y_test,predictions)
  f1_score = metrics.f1_score(y_test,predictions)
  recall_score=metrics.recall_score(y_test,predictions)
  precision_score=metrics.precision_score(y_test,predictions)
  cm1 = confusion_matrix(y_test,predictions)
  scores=abs(cm1[0,1]-cm1[1,0])
  return train,test,cm1,scores,acc_score,f1_score,recall_score,precision_score
```

```
#Removing URLs
def remove_urls(text):
	url_pattern = r'https?://\S+|www\.\S+'
	without_urls = re.sub(pattern=url_pattern, repl='', string=text)
	return without_urls

# Implementation of emoji removing
def remove_emojis(text):
	"""
	Result :- string without any emojis in it
	Input :- String
	Output :- String
	"""
	emoji_pattern = re.compile("["
                               u"\U0001F600-\U0001F64F"  # emoticons
                               u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                               u"\U0001F680-\U0001F6FF"  # transport & map symbols
                               u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                               u"\U00002500-\U00002BEF"  # chinese char
                               u"\U00002702-\U000027B0"
                               u"\U00002702-\U000027B0"
                               u"\U000024C2-\U0001F251"
                               u"\U0001f926-\U0001f937"
                               u"\U00010000-\U0010ffff"
                               u"\u2640-\u2642"
                               u"\u2600-\u2B55"
                               u"\u200d"
                               u"\u23cf"
                               u"\u23e9"
                               u"\u231a"
                               u"\ufe0f"  # dingbats
                               u"\u3030"
                               "]+", flags=re.UNICODE)

	without_emoji = emoji_pattern.sub(r'',text)
	return without_emoji

#Removing unwanted characters
def clean_data(text):
    text=text.replace('âš ï','')
    text=text.replace('âœ','')
    text=text.replace('&amp;','')
    text=text.replace('â€™','')
    text=text.replace('â€œ','')
    text=text.replace('â€','')
    text=text.replace('ðŸ”¥ðŸš’','')
    text=text.replace('â„ï¸','')
    text=text.replace('â€¼ï¸','')
    text=text.replace('ðŸŽµðŸŽ¶','')
    text=text.replace('ðÿž¶ðÿžµ','')
    text=text.replace('ðÿžµðÿž¶','')
    text=text.replace('ðŸŽ¶ðŸŽµ','')
    text=text.replace('ðŸŽ¶ðŸª•','')
    text=text.replace('ªð','')
    text=text.replace('²ð','')
    text=text.replace('µð','')
    text=text.replace('¹ð','')
    text=text.replace('ºð','')
    text=text.replace('ºðÿ','')
    text=text.replace('¼ð','')
    text=text.replace('½ðÿš','')
    text=text.replace('ðÿ','')
    text=text.replace('ðÿœš','')
    text=text.replace('ðÿš','')
    text=text.replace('ðÿž','')
    text=text.replace('\n','')
    for char in ['²the', '³ð', 'áÿ', 'áÿƒáž', 'áž', 'áž¼ážšážˆáÿ', 'áž½áž', 'áž¾ážˆáÿ', 'ážÿáÿ', 'ážÿáž', 'ážšáž', 'ážšáž¹áž', 'ážšáž¼ážœáž', 'ážšáž¾áž', 'âš', 'âž', 'žƒ', 'ƒï']:
      text=text.replace('char','')
    return text

# Removing Extra Whitespaces
def remove_extra_spaces(text):
	"""
	Return :- string after removing extra whitespaces
	Input :- String
	Output :- String
	"""
	space_pattern = r'\s+'
	without_space = re.sub(pattern=space_pattern, repl=" ", string=text)
	return without_space

# Implementation of removing punctuations using string library
def remove_punctuation(text):
	"""
	Return :- String after removing punctuations
	Input :- String
	Output :- String
	"""
	return text.translate(str.maketrans('', '', punctuation))
```
